# NFL Wide Receivers

Fantasy team owners fill out their teams usually through a draft.  Fnatasy owners attempt to select players who will perform well in real life and amass fantasy points for their team.  Fantasy points are awarded through a variety of scoring philosophies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
nfl17 = pd.read_excel('Data Visual Week 4 Data 1.xlsx')
nfl18 = pd.read_excel('Data Visual Week 4 Data 2.xlsx')

# PPR Scoring
nfl17['FP17'] = nfl17['Rec'] + (0.1 * nfl17['Yds']) + (6 * nfl17['TD']) - (2 * nfl17['Fmb'])
nfl18['FP18'] = nfl18['Rec'] + (0.1 * nfl18['Yds']) + (6 * nfl18['TD']) - (2 * nfl18['Fmb'])
```

```python
nfl17 = nfl17[nfl17['Pos'].str.upper() == 'WR']

nfl17.groupby('Pos').count()
```

```python
nfl18 = nfl18[nfl18['Pos'].str.upper() == 'WR']

nfl18.groupby('Pos').count()
```

```python
data17and18 = pd.merge(nfl17[['Player_Cleaned', 'FP17']], nfl18[['Player_Cleaned', 'FP18']], how="outer", on='Player_Cleaned')
data17and18.fillna(0)
```

```python
data17and18.sort_values(by=['FP17'], inplace=True, ascending=False)
df1 = data17and18.reset_index()
Top17 = df1.iloc[39].FP17
Top17
```

```python
data17and18.sort_values(by=['FP18'], inplace=True, ascending=False)
df2 = data17and18.reset_index().head(40)
Top18 = df2.iloc[39].FP18
Top18
```

```python
data17and18['Color'] = np.where((data17and18['FP17'] >= Top17) & (data17and18['FP18'] >= Top18), 'green', 
                                np.where((data17and18['FP17'] >= Top17) & (data17and18['FP18'] < Top18), 'red',
                                        np.where((data17and18['FP17'] < Top17) & (data17and18['FP18'] >= Top18), 'yellow',
                                                'blue')))
```

```python
data17and18[data17and18['Color'] == 'green']
```

```python
plt.scatter(data17and18['FP17'], data17and18['FP18'])

ax = plt.axes()

plt.axis('equal')

#ax.plot([0, 1], [0, 1], transform=ax.transAxes, color='orange')
ax.axhline(y=Top18, linestyle='--', color='green')
ax.axvline(x=Top17, linestyle='--', color='red')
ax.set_facecolor('white')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('2017 Fantasy Points')
plt.ylabel('2018 Fantasy Points')
plt.title('Wide Receiver Fantasy Points Comparison')

plt.text(300, 190, 'Consistent Top WR')
plt.text(300, 50, 'Fallen Top WR')
plt.text(-50, 190, 'Ascending Top WR')

plt.savefig('Week 4.pdf')
```
